#### def

```python
def test_classification(oneGroup):
    oneGroup = groupDf.get_group(list(groupDf.groups)[1])
    eachGroup = oneGroup.reset_index(drop=True)

    from sklearn.model_selection import train_test_split

    features = ['OpportunityNumber','SuppliesSubgroup', 'SuppliesGroup','RouteToMarket', 'ElapsedDaysInSalesStage',\
               'SalesStageChangeCount', 'TotalDaysIdentifiedThroughClosing','TotalDaysIdentifiedThroughQualified', 'OpportunityAmountUSD',\
               'ClientSizeByRevenue', 'ClientSizeByEmployeeCount','RevenueFromClientPastTwoYears', 'CompetitorType',\
               'RatioDaysIdentifiedToTotalDays', 'RatioDaysValidatedToTotalDays','RatioDaysQualifiedToTotalDays', 'DealSizeCategory']
    label = ['OpportunityResult']

    trainingData_features,\
    testData_features,\
    trainingData_label,\
    testData_label= train_test_split(eachGroup[features],
                    eachGroup[label],
                    test_size = 0.3, 
                     random_state = 3)

    from sklearn import svm

    model_method = svm.SVC() #정확도 높게 나오는거 c...로 어떻게?

    model = model_method.fit(trainingData_features,
                            trainingData_label)#학습

    predict = model.predict(testData_features)
    predict

    testData_features2 = testData_features.reset_index(drop=True)

    testData_label2 = testData_label.reset_index(drop=True)

    finalResult = pd.concat(\
                           [testData_features2,testData_label2],\
                           axis=1)

    finalResult['PREDICT'] = predict
    
    return finalResult
```

```python
groupKey=[ 'Region']
```

```python
testResult = data.groupby(groupKey).apply(test_classification)
```

```python
testResult.head()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
as_score = accuracy_score(testResult.OpportunityResult,testResult.PREDICT)
```

```python
as_score
```

```python
clfc_report = classification_report(testResult.OpportunityResult,testResult.PREDICT)
```

```python
print("result= \n",clfc_report)
```
