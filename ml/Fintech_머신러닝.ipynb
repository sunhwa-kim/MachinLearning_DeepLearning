```python
def mlFcst(oneGroup):
    # data2.reset_index(drop=True,inplace=True)

    # groupData = data2.groupby(groupKey)

    # oneGroup = groupData.get_group(  list(groupData.groups)[8])
    # len(oneGroup)
# features 너무 적음....
    oneGroup.sort_values(by=['YEARWEEK'],inplace=True)

    oneGroup.reset_index(drop=True,inplace=True)

    standardIndex =round(len(oneGroup)*0.8)
    standardIndex

    oneGroup.sort_values(by='YEARWEEK',inplace=True)

    standardYearweek = oneGroup.loc[standardIndex,['YEARWEEK']].values[0]

    trainingData = oneGroup[oneGroup.YEARWEEK <= standardYearweek]

    testData =  oneGroup[oneGroup.YEARWEEK > standardYearweek]

    trainingData.reset_index(drop=True,inplace=True)
    testData_02 = testData.reset_index()

    trainingData_features = trainingData[['PRO_PERCENT','HOLY_ENCO','PTOMO_ENCO']]
    trainingData_label = trainingData[['QTY']]
    testData_features = testData_02[['PRO_PERCENT','HOLY_ENCO','PTOMO_ENCO']]
    testData_label = testData_02[['QTY']]

    model_method = tree.DecisionTreeRegressor()

    model = model_method.fit(trainingData_features,trainingData_label)

    predict = model.predict(testData_features)

    predic_df = pd.DataFrame(data=predict, columns=['PREDICT'])

    testData['PREDICT'] = predic_df

    return testData
```

```python
resultData = data2.groupby(groupKey).apply(mlFcst).reset_index(drop=True)
```
